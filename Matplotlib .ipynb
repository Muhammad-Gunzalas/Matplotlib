### Grid

```python
y=[1,4,55,6,17,2,33,4]
x1=[1,12,13,5,16,77,88,99]
x2=[1,4,5,6,13,16,19,34]
plt.plot(x1, y,'ys-')
plt.plot(x2, y,'g*--')
plt.legend(labels=('TV','Smartphone'),loc='lower right')

plt.grid(True)
plt.title('AD effects on sales')
plt.xlabel('medium')
plt.ylabel('sales')
plt.show()
```

```python
menmean=[20,35,30,27,40]
womenmean=[25,12,34,20,25]
group=np.arange(5)

plt.bar(group,  menmean, width=.35,color='r')
plt.bar(group,womenmean, width=.35,  bottom=menmean ,color='b')

plt.xlabel('scores')
plt.ylabel('Groups')
plt.title('Scores by groups and gender')

plt.xticks(group,('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0,100,10))
plt.legend(labels=['Men','Women'])
plt.show()
```

```python
lans=np.array(['C','JAva','C++','Python','php'])
students=np.array([23,17,34,12,3])
group=np.arange(5)

plt.bar(group,lans,width=.40,color='g')
plt.bar(group,students,width=.40,bottom=lans,color='r')

plt.xlabel('Languages')
plt.ylabel('Students')

plt.xticks(group,(lans))
plt.yticks(np.arange(0,20,5))
plt.legend(labels=['lans','students'])


plt.show()
```

### 3D plotting

```python
from mpl_toolkits import mplot3d
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(0,1,100)
x=z*np.sin(20*2)
y=z*np.cos(20*2)

ax.plot3D(x,y,z,'gray')
plt.title('3d line plot')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(0,1,100)
x=z*np.sin(20*2)
y=z*np.cos(20*2)
c=x+y

ax.scatter(x,y,z,c=c)
plt.title('3d line plot')
plt.show()
```
